**Generate 3 testnet addresses (address_1, address_2, address_3; should
start with either m or n)**

```python
import random

def generate_wallet_name():
    num = random.randrange(10**9)
    print(f"Generated Wallet_{num}")
    return f'Wallet_{num}'
```

```python
from bitcoinlib.mnemonic import Mnemonic

mnemonic = Mnemonic()

passphrases = []
for i in range(3):
    passphrase = mnemonic.generate()
    print(f"Passphrase {i}:")
    print(passphrase)
    passphrases.append(passphrase)
```

My passphrases:
1) member history outside quarter parade dawn fitness secret sight absurd dad final
2) west brand physical alien monitor canyon tower reduce step book enemy flame
3) fade coyote mother car couch castle cannon proud this green wave comic

```python
from bitcoinlib.keys import *

hd_keys = []
for passphrase in passphrases:
    hd_key = HDKey.from_passphrase(passphrase, network = 'testnet')
    hd_keys.append(hd_key)
```

```python
for key in hd_keys:
    print(key.info())
```

```python
from bitcoinlib.wallets import *

wallets = []
for i in range(3):
    wallet = wallet_create_or_open(generate_wallet_name(), keys=hd_keys[i], network = 'testnet')
    wallets.append(wallet)
```

```python
for w in wallets:
    print(w.get_key().address)
```

**1. (1 point) Calculate double Hash_256 of your name. Take two inputs from
the user, i.e., first name and last name. Concatenate string and take
SHA256(SHA256(<first name><lastname>)).**

```python
from hashlib import sha256

my_name = "Deniz Ozmaden"
hashed = sha256(bytes(my_name, encoding='utf-8'))
print(f"First hash: {hashed.hexdigest()}")

hashed_twice = sha256(hashed.digest()).hexdigest()
print(f"Second hash: {hashed_twice}")
```
