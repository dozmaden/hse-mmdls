**1. (1 point) Create a user-interacting interface to connect to different blockchain networks, i.e. main net, testnets (any two), and ganache**

```python
from web3 import Web3, HTTPProvider

networks = {
  "ethereum_mainnet": "https://mainnet.infura.io/v3/9ccda88db66842758b856e6a5be54e54",
  "ethereum_gorli": "https://goerli.infura.io/v3/9ccda88db66842758b856e6a5be54e54",
  "ethereum_sepolia": "https://sepolia.infura.io/v3/9ccda88db66842758b856e6a5be54e54",
}

w3 = None

while(w3 == None): 
    print("Please enter the name of the network you are willing to connect to: ")
    print("(ethereum_mainnet, ethereum_gorli, ethereum_sepolia, ganache)")
    network_input = input()
    while (network_input == "ganache"):
        print("Enter RPC server: ")
        server = input()
        w3 = Web3(HTTPProvider(server))
    w3 = Web3(HTTPProvider(networks[network_input]))

print("Latest Ethereum block number" , w3.eth.block_number)
```

**4. (1 point) Send some Eth test coins from address_1 to address_2 and provide Transaction_ID.**

```python
from web3.gas_strategies.rpc import rpc_gas_price_strategy

w3.eth.set_gas_price_strategy(rpc_gas_price_strategy)

tx_create = w3.eth.account.sign_transaction(
    {
        "nonce": w3.eth.get_transaction_count(first_account.address),
        "gasPrice": w3.eth.generate_gas_price(),
        "gas": 21000,
        "to": second_account.address,
        "value": w3.toWei("0.1", "ether"),
    },
    first_account.privateKey,
)

tx_hash = w3.eth.send_raw_transaction(tx_create.rawTransaction)
tx_receipt = w3.eth.wait_for_transaction_receipt(tx_hash)

print(f"Sent! Transaction hash: { tx_receipt.transactionHash.hex() }")
```
